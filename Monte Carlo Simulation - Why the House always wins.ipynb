Game such that you roll 1-51 and the house wins. You roll anything between 52-100 and we win. Hence the house edge will be about 2%. 
Let's say we bet $1 in the game. 
Then expected profit for player = (49/100)*1 - (51/100)*1 = -0.02
2% profit in the favor of the house. 

```python
#import the libraries
import random
import matplotlib.pyplot as plt
```

```python
#creating a function to simulate a die roll. 
def die_roll():
    
    dice = random.randint(1,100)
    
    if dice <= 51:
        return False
    else:
        return True
```

```python
#creating a function that simulates the bets
#total_funds, total_wager, number_of_time_game_is_played

def play(total_funds, wager_amount, total_plays):
    play_num = []
    funds = []
    final_funds = []
    
    play = 1
    while play < total_plays:
        if die_roll():
            total_funds += wager_amount
            play_num.append(play)
            funds.append(total_funds)
        else:
            total_funds -= wager_amount
            play_num.append(play)
            funds.append(total_funds)
            
        play += 1
    
    plt.plot(play_num, funds)
    final_funds.append(funds[-1])
    return (final_funds)
```

```python
#calling the above function in several scenarios and runnning each scenario a 100 times
scen = 1
final_funds1 = []

while scen <=100:
    end_fund = play(10000,100,5)
    scen += 1
    
plt.ylabel('Player money in $')
plt.xlabel('Number of bets')
plt.show()
    
print('The player starts with $10000 and ends with $' + str(sum(end_fund)/len(end_fund)))
```
